```python
import requests
from bs4 import BeautifulSoup
from lxml import etree
from requests.exceptions import HTTPError

lista_titulos = []
lista_km = []
lista_anio = []
lista_precio = []
lista_moneda = []
lista_url = []

Siguiente = 'https://autos.mercadolibre.com.ar/_PriceRange_{}USD-{}USD'

price_min = 1000
price_max = 1500

print('Obteniendo artículos entre', price_min, 'USD y', price_max, 'USD')

while price_max <= 1000000:
    Siguiente = Siguiente.format(price_min, price_max)
    
    try:
        r = requests.get(Siguiente)
        
        try:
            
            r.raise_for_status()  # Comprueba si hay errores en la solicitud HTTP
        
        except HTTPError:

            price_min += 501
            price_max += 501
            Siguiente = 'https://autos.mercadolibre.com.ar/_PriceRange_{}USD-{}USD'
            print('Obteniendo artículos entre', price_min, 'USD y', price_max, 'USD')
        
        soup = BeautifulSoup(r.content, 'html.parser')
        dom = etree.HTML(str(soup))
        
        # Títulos
        titulo = soup.find_all('h2', attrs={'class': 'ui-search-item__title'})
        titulos = [i.text for i in titulo]
        lista_titulos.extend(titulos)

        anio = dom.xpath('//div[@class="ui-search-item__group ui-search-item__group--attributes"]//li[1]')
        anio = [i.text for i in anio]
        lista_anio.extend(anio)

        km = dom.xpath('//div[@class="ui-search-item__group ui-search-item__group--attributes"]//li[2]')
        km = [i.text for i in km]
        lista_km.extend(km)

        precio = dom.xpath('//div[@class="ui-search-price__second-line"]//span//span[2]')
        precio = [i.text for i in precio]
        lista_precio.extend(precio)

        moneda = dom.xpath('//div[@class="ui-search-price__second-line"]//span//span[1]')
        moneda = [i.text for i in moneda]
        lista_moneda.extend(moneda)

        div_tags = soup.find_all('div', class_='ui-search-item__group ui-search-item__group--title')

        # Encontrar todas las etiquetas <a> dentro de las etiquetas <div>
        for div_tag in div_tags:
            link_tag = div_tag.find('a')
            # Obtener la URL de la etiqueta <a>
            if link_tag:
                url_extraida = link_tag.get('href')
                lista_url.append(url_extraida)

        Ini = soup.find('span', attrs={'class': 'andes-pagination__link'}).text
        Ini = int(Ini)
        Cant = int(soup.find('li', attrs={'class': 'andes-pagination__page-count'}).text.split(' ')[1])

        print('Página', Ini, 'de', Cant)

        if Ini != Cant:
            Siguiente = dom.xpath(
                '//div[@class="ui-search-pagination"]//nav//li[@class="andes-pagination__button andes-pagination__button--next"]/a')[0].get('href')
        else:
            price_min += 501
            price_max += 501
            Siguiente = 'https://autos.mercadolibre.com.ar/_PriceRange_{}USD-{}USD'
            print('Obteniendo artículos entre', price_min, 'USD y', price_max, 'USD')

    except AttributeError:
        
        price_min += 501
        price_max += 501
        Siguiente = 'https://autos.mercadolibre.com.ar/_PriceRange_{}USD-{}USD'
        print('Obteniendo artículos entre', price_min, 'USD y', price_max, 'USD')

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Obtener la página principal con la lista de URLs
r = requests.get('https://autos.mercadolibre.com.ar/audi/audi_PriceRange_10000USD-53000USD')
soup = BeautifulSoup(r.content, 'html.parser')

# Encontrar todas las URLs de los productos
urls = soup.find_all('a', attrs={'class': 'ui-search-item__group__element'})
urls = [i.get('href') for i in urls]

data = {'Marca': [], 'Modelo': [], 'Color': [], 'Combustible': [], 'Km':[], 'Carroceria':[], 'Version':[], 'Url':url}

for url in urls:
    auto_data = {'Marca': None, 'Modelo': None, 'Color': None, 'Combustible': None, 'Km':None, 'Carroceria':None, 'Version':None}

    # Hacer una solicitud a la URL del producto
    r = requests.get(url)
    soup = BeautifulSoup(r.content, 'html.parser')

    # Encontrar todos los divs con la clase andes-table__header__container
    divs = soup.find_all('div', attrs={'class': 'andes-table__header__container'})

    # Encontrar todos los spans con la clase andes-table__column--value
    spans = soup.find_all('span', attrs={'class': 'andes-table__column--value'})

    for div, span in zip(divs, spans):
        div_text = div.text.strip()
        span_text = span.text.strip()

        if div_text == 'Marca':
            auto_data['Marca'] = span_text
        elif div_text == 'Modelo':
            auto_data['Modelo'] = span_text
        elif div_text == 'Color':
            auto_data['Color'] = span_text
        elif div_text == 'Tipo de combustible':
            auto_data['Combustible'] = span_text
        elif div_text == 'Tipo de carrocería':
            auto_data['Carroceria'] = span_text
        elif div_text == 'Kilómetros':
            auto_data['Km'] = span_text
        elif div_text == 'Versión':
            auto_data['Version'] = span_text

    data['Marca'].append(auto_data['Marca'])
    data['Modelo'].append(auto_data['Modelo'])
    data['Color'].append(auto_data['Color'])
    data['Combustible'].append(auto_data['Combustible'])
    data['Carroceria'].append(auto_data['Carroceria'])
    data['Version'].append(auto_data['Version'])
    data['Km'].append(auto_data['Km'])

# Convierte los datos en un DataFrame de pandas
df = pd.DataFrame(data)
```
